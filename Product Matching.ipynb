### DataFrames

```python
Flipkart_data = pd.DataFrame(data=df,columns=["product_name",'retail_price','discounted_price'])
```

```python
Flipkart_data #DataFrame of neccesary columns
```

```python
Amazon_data=pd.DataFrame(data=Amazon,columns=["product_name",'retail_price','discounted_price']) #DataFrame of neccesary columns
```

```python
Amazon_data
```

```python
Flipkart_Table=Flipkart_data.rename(columns={"product_name":'Product Name in Flipkart','retail_price':'Retail Price in Flipkart','discounted_price':'Discounted Price in Flipkart'}) #renames of columns
```

```python
Flipkart_Table.head()
```

```python
Amazon_Table=Amazon_data.rename(columns={"product_name":'Product Name in Amazon','retail_price':'Retail Price in Amazon','discounted_price':'Discounted Price in Amazon'}) #renames of columns
```

```python
Amazon_Table.head()
```

### Final Table

```python
Final_Table=pd.concat([Flipkart_Table,Amazon_Table],axis=1,join='inner') # show final output with inner join
```
